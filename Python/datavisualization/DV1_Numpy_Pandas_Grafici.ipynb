## Programmazione Orientata agli Array con NumPy

La libreria **NumPy** (Numerical Python) è l'implementazione Python preferita per lavorare con gli Array.
- Offre un tipo di array n-dimensionale ad alte prestazioni chiamato `ndarray`
- Le operazioni sugli array sono fino a due ordini di grandezza più veloci di quelle sulle liste.
- Oltre $450$ librerie Python dipendono da NumPy. Molte popolari librerie di data science come Pandas, SciPy (Scientific Python) e Keras (per il deep learning) sono basate o dipendono da NumPy.

Il punto di forza di NumPy è la ''programmazione orientata agli array'', che utilizza la programmazione in stile funzionale con iterazione interna per rendere le manipolazioni degli array concise e dirette, eliminando i tipi di bug che possono verificarsi con l'iterazione esterna di cicli programmati esplicitamente.

#### Disegnare un Plot da uno script

Se stai usando Matplotlib dall'interno di uno script, la funzione ``plt.show()`` avvia un ciclo di eventi, cerca tutti gli oggetti figura attualmente attivi e apre una o più finestre interattive che mostrano la tua figura o le tue figure.

Quindi, ad esempio, potresti avere un file chiamato *myplot.py* contenente quanto segue:

```python
# ------- file: myplot.py ------
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

È quindi possibile eseguire questo script dal prompt della riga di comando, che si tradurrà in una finestra che si apre con la figura visualizzata:

```
$ python myplot.py
```

Il comando ``plt.show()`` nasconde molto bene i dettagli implementativi, poiché deve interagire con il backend grafico interattivo del tuo sistema.
I dettagli di questa operazione possono variare notevolmente da sistema a sistema e persino da installazione a installazione.

### Due interfacce al prezzo di una

Una caratteristica potenzialmente fonte di confusione di Matplotlib è la sua doppia interfaccia: una comoda interfaccia basata sullo stato in stile MATLAB e un'interfaccia orientata agli oggetti più potente.

#### Interfaccia in stile MATLAB 

Matplotlib è stato originariamente scritto come alternativa a Python per gli utenti MATLAB e gran parte della sua sintassi riflette questo fatto.
Gli strumenti in stile MATLAB sono contenuti nell'interfaccia pyplot (`` plt``).
Ad esempio, il codice seguente sarà probabilmente abbastanza familiare agli utenti MATLAB:

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(0, 10, 100)
````

````python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
````

#### Interfaccia orientata agli oggetti

L'interfaccia orientata agli oggetti è disponibile per situazioni più complicate e per quando si desidera un maggiore controllo sulla propria figura.
Piuttosto che dipendere da una qualche nozione di una figura o assi "attivi", nell'interfaccia orientata agli oggetti le funzioni di plottaggio sono *metodi* di oggetti espliciti ``Figure`` e ``Axes``.
Per ricreare la figura precedente utilizzando questo stile di plot, si può fare quanto segue:

````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
````

### Regolazione del grafico: colori e stili delle linee

La prima regolazione che potresti voler apportare a un grafico è controllare i colori e gli stili delle linee.
La funzione ``plt.plot()`` accetta argomenti aggiuntivi che possono essere utilizzati per specificarli.
Per regolare il colore, puoi usare la parola chiave ``color``, che accetta un argomento stringa che rappresenta praticamente qualsiasi colore immaginabile.
Il colore può essere specificato in vari modi:

````python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
````

Se non viene specificato alcun colore, Matplotlib scorrerà automaticamente attraverso una serie di colori predefiniti per più righe.

Allo stesso modo, lo stile della linea può essere regolato utilizzando la parola chiave ``linestyle``:

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
````

Se vuoi essere estremamente conciso, questi codici ``linestyle`` e ``color`` possono essere combinati in un singolo argomento non parola chiave per la funzione ``plt.plot()``:

````python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
````

Questi codici colore a carattere singolo riflettono le abbreviazioni standard nei sistemi di colore RGB (Red/Green/Blue) e CMYK (Cyan/Magenta/Yellow/blacK), comunemente usati per la grafica a colori digitale.

### Regolazione del grafico: limiti degli assi

Il modo più semplice per regolare i limiti degli assi è usare i metodi ``plt.xlim()`` e ``plt.ylim()``:

````python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
````

Un utile metodo correlato è ``plt.axis()`` (da notare qui la potenziale confusione tra *axes* con *e* e *axis* con *i*).
Il metodo ``plt.axis()`` consente di impostare i limiti ``x`` e ``y`` con una singola chiamata, passando una lista che specifica ``[xmin, xmax, ymin, ymax]``:

````python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
````

Il metodo ``plt.axis()`` va anche oltre, permettendo di fare cose come stringere automaticamente i limiti attorno alla figura corrente:

````python
plt.plot(x, np.sin(x))
plt.axis('tight');
````

Consente specifiche di livello ancora più elevato, come garantire un rapporto di aspetto uguale in modo che sullo schermo, un'unità in ``x`` sia uguale a un'unità in ``y``:

````python
plt.plot(x, np.sin(x))
plt.axis('equal');
````

### Etichettatura dei grafici

I titoli e le etichette degli assi sono le più semplici:

````python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
````

Quando vengono visualizzate più linee all'interno di un singolo asse, può essere utile creare una legenda del grafico che etichetta ogni tipo di linea.

````python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
````

## Visualizzare gli Errori

Per qualsiasi misurazione scientifica, tener conto in maniera accurata degli errori è quasi altrettanto importante, se non più importante, della misurazione accurata del numero stesso.

Ad esempio, si immagini che io stia usando alcune osservazioni astrofisiche per stimare la costante di Hubble, la misura locale del tasso di espansione dell'Universo.
So che la letteratura attuale suggerisce un valore di circa 71 (km/s)/Mpc, e misuro un valore di 74 (km/s)/Mpc con il mio metodo. 
I valori sono coerenti? L'unica risposta corretta, data questa informazione, è questa: non c'è modo di saperlo.

Supponiamo che io aumenti queste informazioni con le incertezze riportate: la letteratura corrente suggerisce un valore di circa 71 $\pm$ 2.5 (km/s)/Mpc e il mio metodo ha misurato un valore di 74 $\pm$ 5 (km/s)/Mpc. Ora i valori sono coerenti? Questa è una domanda a cui è possibile rispondere quantitativamente.

Nella visualizzazione di dati e risultati, mostrare questi errori in modo efficace può far sì che un grafico trasmetta informazioni molto più complete.

È possibile creare una errorbar di base con una singola chiamata alla funzione Matplotlib:

````python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
````

### Istogrammi bidimensionali e binnings

Proprio come creiamo istogrammi in una dimensione dividendo la linea numerica in contenitori, possiamo anche creare istogrammi in due dimensioni dividendo punti tra contenitori bidimensionali.

Si considerino due array ``x`` e ``y`` tratti da una distribuzione gaussiana multivariata:

````python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
````

## Esplorazione di Plot in Seaborn 

````python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], density=True, alpha=0.5)
````

Piuttosto che un istogramma, possiamo ottenere una stima uniforme della distribuzione utilizzando una stima della densità del kernel, cosa che Seaborn fa con ``sns.kdeplot``:

````python
sns.kdeplot(data=data, x=data.x, y=data.y);
````

````python
sns.axes_style('white')
sns.jointplot(x="x", y="y", data=data, kind='kde');
````
