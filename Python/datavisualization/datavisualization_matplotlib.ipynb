```python
#creazione grafico

import matplotlib.pyplot as plt

plt.figure(figsize=(4,3)) #imposta la dimensione del grafico
plt.plot([1,2,3],[3,3,2]) #disegna la linea 2D
plt.title("Grafico d'esempio") #imposta il titolo in alto
plt.show() #mostra tutto il grafico

```

```python
#subplot 

import matplotlib.pyplot as plt

fig , ax = plt.subplots(2,3,figsize = (6,3)) #crea due file di 3 grafici,larghi 6 e alti 3
fig.tight_layout() #gestisce la rappresentazione grafica corretta
plt.show() 
```

```python
import numpy as np


fig, ax = plt.subplots(3,3,figsize=(8,6))
for i in range(3):
    for j in range(3):
        x = np.linspace(0,10,10)
        y = np.linspace(0,10,10) + np.random.randint(-2,2,10)
        ax[i][j].set_title(f"plot posizione{i},{j}",fontsize=10,y=1.05)
        ax[i][j].set_ylabel("asse y")
        ax[i][j].set_xlabel("asse x")
        ax[i][j].set_ylabel("asse y")
        ax[i][j].plot(x,y)
        
fig.tight_layout()
fig.suptitle("titolo del grafico",fontsize = 16,y = 1.05)
plt.show()
```

```python
h_min = [10,2,8]
h_max = [8,5,3]

x = np.arange(3)
labels =['Sensor1','Sensor2','Sensor3']

width = 0.4

fig,ax = plt.subplots(figsize = (3,2))
ax.bar(x+width/2,h_min,width=width)
ax.bar(x-width/2,h_max,width=width)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc = (1.1,0.5))
plt.show()

fig.savefig('./bar.png')
```
